Section 3 - Cleaning

-For this analysis we are using only columns for EDA + insights
-Column names were standardized using snake_case and only core explanatory variables were retained in order to focus on interpretable relationships.

```python
df = df.rename(columns={
    'Country name':'country',
    'Regional indicator':'region',
    'Ladder score': 'happiness_score',
    'Logged GDP per capita': 'gdp_per_capita',
    'Social support': 'social_support',
    'Healthy life expectancy': 'life_expectancy',
    'Freedom to make life choices': 'freedom',
    'Generosity': 'generosity',
    'Perceptions of corruption': 'corruption_perception'
})
```

```python
cols_to_drop = [
    'Standard error of ladder score',
    'upperwhisker',
    'lowerwhisker',
    'Ladder score in Dystopia',
    'Explained by: Log GDP per capita',
    'Explained by: Social support',
    'Explained by: Healthy life expectancy',
    'Explained by: Freedom to make life choices',
    'Explained by: Generosity',
    'Explained by: Perceptions of corruption',
    'Dystopia + residual'
]

df = df.drop(columns=cols_to_drop)
```

```python
plt.figure()
sns.histplot(df['happiness_score'], kde=True)
plt.title('Distribution of Happiness Score (2021)')
plt.show()
```

```python
df['happiness_score'].describe()
```

```python
sns.countplot(df["region"])
plt.xticks(rotation = 60)
plt.show()
```

```python
plt.figure(figsize=(15, 8))

regions = df["region"].unique()

for region in regions:
    sns.kdeplot(
        data=df[df["region"] == region],
        x="happiness_score",
        fill=True,
        alpha=0.3,
        linewidth=2,
        common_norm=False,
        label=region
    )

plt.axvline(
    df["happiness_score"].mean(),
    color="black",
    linestyle="--",
    label="Global Mean"
)

plt.title("Distribution of Happiness Score by Region (2021)")
plt.xlabel("Happiness Score")
plt.ylabel("Density")
plt.legend(loc='upper left')
plt.show()
```

```python
corr = df.select_dtypes(include='number').corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
europe = df[df["region"].str.contains("Europe", na=False)]["happiness_score"]
rest = df[~df["region"].str.contains("Europe", na=False)]["happiness_score"]

europe_mean = europe.mean()
rest_mean = rest.mean()

europe_mean, rest_mean
```
